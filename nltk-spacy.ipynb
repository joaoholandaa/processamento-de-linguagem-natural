### NLTK - TokenizaĂ§ĂŁo - Twitter (X)

```python
from nltk.tokenize import TweetTokenizer
texto2 = "I'm very veryyyy happyyyyyyy #betterlife @barneys :P :D"
```

Analizando a diferenĂ§a entre o word_tokenize e o twitter tokenize

```python
word_tokenize(texto2)
```

```python
twitterTokenizer = TweetTokenizer()
twitterTokenizer.tokenize(texto2)
```

```python
twitterTokenizer2 = TweetTokenizer(strip_handles=True, reduce_len=True) #strip_handles: remove os nomes de usuĂĄrio do texto; reduce_len=True: substitui sequĂȘncias de 3 ou mais caracteres repetidos por sequĂȘncia de 3 caracteres
twitterTokenizer2.tokenize(texto2)
```

### Spacy - TokenizaĂ§ĂŁo

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

Gerando um objeto o qual gerarĂĄ um objeto iterĂĄvel e cada item representa um token, e cada token contĂ©m alĂ©m da palavra em si, sua versĂŁo simplificada, funĂ§ĂŁo sintĂĄtica, etc.

```python
sentenca = nlp('JoĂŁo Pedro is working at Google in the South America. He works very hard :D')
for token in sentenca: #o atributo .text retorna o texto do respectivo token
  print(token.text)
```

```python
#Caso queira considerar primeiramente cada sentenĂ§a como token tambĂ©m Ă© possĂ­vel.
#Para isso basta utilizar a propriedade sents do objeto gerado
doc = nlp('JoĂŁo Pedro is working at Google in the South America. She works very hard!')
```

```python
for sent in doc.sents:
  print(sent.text)
```
